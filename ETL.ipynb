# Hospital capacity in each county data

```python
US_hosp = pd.read_json('Data/usa-hospital-beds.geojson', lines=True)
US_hosp.head()
```

```python
US_hosp.count()
```

__removing PR as we removed it in the US_pop_c__

```python
US_hosp_c = US_hosp.drop(['OBJECTID', 'HOSPITAL_TYPE' , 'HQ_ADDRESS', 'HQ_ADDRESS1', 'HQ_CITY', 'HQ_ZIP_CODE', 'STATE_NAME' , 'STATE_FIPS' , 'CNTY_FIPS', 'FIPS', 'latitude', 'longtitude'], axis = 1).reset_index(drop=True)
US_hosp_c = US_hosp_c[US_hosp_c['HQ_STATE'] != 'PR']
US_hosp_c.reset_index(drop=True)
```

__joining the hospital data with US_pop_c using county names__

```python
US_hosp_pop = US_hosp_c.join(US_pop_c.set_index(['County', 'Abbreviation']), on = ['COUNTY_NAME', 'HQ_STATE']).drop('State', axis = 1).reset_index(drop = True)
US_hosp_pop.count()
```

__There are 110 counties in the hospital data that we couldn't find a match for it in the US_pop_c__

```python
US_hosp_pop[US_hosp_pop['Population Estimate 2018'].isna()].groupby(by = ['COUNTY_NAME', 'HQ_STATE']).count().index.size
```

__Let's get these 110 counties__

```python
temp = US_hosp_pop[US_hosp_pop['Population Estimate 2018'].isna()].groupby(by = ['COUNTY_NAME', 'HQ_STATE']).count().index.tolist()
temp
```

__This is a brute force way to find similar names of the 110 counties in the US_pop_c data and then replace them in the hosptial data__

```python
count_not_found = 0
new_counties = []
old_counties = []
for i in temp:
    old_counties.append(i[0])
    
    new_county = US_pop_c[US_pop_c['County'].str.contains(i[0]) & US_pop_c['Abbreviation'].str.contains(i[1])]['County'].values
    if not new_county.any(): 
        new_county = US_pop_c[US_pop_c['County'].str.contains(i[0].replace(' County', '')) & US_pop_c['Abbreviation'].str.contains(i[1])]['County'].values
        if not new_county.any():    
            new_county = US_pop_c[US_pop_c['County'].str.contains(i[0].replace(' County', '').capitalize()) & US_pop_c['Abbreviation'].str.contains(i[1])]['County'].values
            if not new_county.any():
                new_county = US_pop_c[US_pop_c['County'].str.contains(i[0].replace(' City', ' city')) & US_pop_c['Abbreviation'].str.contains(i[1])]['County'].values
                if not new_county.any():
                    new_county = US_pop_c[US_pop_c['County'].str.contains(i[0].replace(' ', '')) & US_pop_c['Abbreviation'].str.contains(i[1])]['County'].values
    
    
    if new_county.any(): 
        new_county_name = new_county[0]
        new_counties.append(new_county_name)
        US_hosp_pop[(US_hosp_pop['COUNTY_NAME'] == i[0]) & (US_hosp_pop['HQ_STATE'] == i[1])] = US_hosp_pop[(US_hosp_pop['COUNTY_NAME'] == i[0]) & (US_hosp_pop['HQ_STATE'] == i[1])].replace({'COUNTY_NAME' : {i[0] : new_county_name}})
    else:
        new_county_name = '>>>>>>>>>>> NOT FOUND: ' + i[0] + ' ' + i[1]
        count_not_found += 1
    print(i[0] + ' >>>>> ' + new_county_name)
print(count_not_found)
print(len(new_counties))
print(len(old_counties))
```

__now we join them again, we see that we get population data for all the counties. Note: there are hosptials without known county, so we will drop thoses__

```python
US_hosp_pop.drop(['Population Estimate 2018', 'Id', 'Id2'], axis = 1, inplace = True)
US_hosp_pop_c = US_hosp_pop.join(US_pop_c.set_index(['County', 'Abbreviation']), on = ['COUNTY_NAME', 'HQ_STATE']).drop('State', axis = 1).reset_index(drop = True)
US_hosp_pop_c.count()
```

__Final check to see if we have any missing values__

```python
US_hosp_pop_c = US_hosp_pop_c[US_hosp_pop_c['Population Estimate 2018'].notna()]
US_hosp_pop_c.fillna(value = 0, inplace = True)
US_hosp_pop_c.count()
```

__Here we want to get the total number of beds, icu, etc... in each county, so we Sum() over the counties__

```python
temp1 = US_hosp_pop_c.groupby(by=['Id', 'Id2', 'Population Estimate 2018']).sum().reset_index()
temp1
```

__Here we want to get the total number of hospitals in each county__

```python
temp2 = US_hosp_pop_c.groupby(by='Id').count()[['HOSPITAL_NAME']].reset_index()
temp2
```

__We join the two previos temp dataframes to get hospital capacity data for each county because we don't care about the names of the hopsitals, etc.. NOTE: Here we only have data about hospital capacity in only 2481 counties instead of the 3142 counties in US_pop_c__

```python
US_hosp_pop_c = temp1.join(temp2.set_index('Id'), on = 'Id').rename(columns = {'HOSPITAL_NAME' : 'NUM_HOSPITALS'})
US_hosp_pop_c
```

__Correcting the bed utilization which is a percentage__

```python
US_hosp_pop_c['BED_UTILIZATION'] = US_hosp_pop_c['BED_UTILIZATION']/US_hosp_pop_c['NUM_HOSPITALS']
US_hosp_pop_c
```

__Unfortunately there are cases in unknown counties but known states__

```python
US_cases_c[US_cases_c['county'] == 'Unknown']
```

__Finding those states__

```python
temp = US_cases_c[US_cases_c['county'] == 'Unknown']['Abbreviation'].unique().tolist()
temp
```

__To solve the problem of the missing county names, we make an assumption: we assume that the Covid cases happened in the counties of largest population in the corresponding state. This can be justified by the method in which Covid spread__

```python
for i in temp:
    set_county = US_pop_c[US_pop_c['Abbreviation'] == i].sort_values(by=['Population Estimate 2018'], ascending = False)['County'].values[0]
    US_cases_c[(US_cases_c['Abbreviation'] == i) & (US_cases_c['county'] == 'Unknown')] = US_cases_c[(US_cases_c['Abbreviation'] == i) & (US_cases_c['county'] == 'Unknown')].replace({'county' : {'Unknown' : set_county}})
```

```python
US_cases_c[US_cases_c['county'] == 'Doña Ana'] = US_cases_c[US_cases_c['county'] == 'Doña Ana'].replace({'county' : {'Doña Ana' : 'Dona Ana'}})
```
